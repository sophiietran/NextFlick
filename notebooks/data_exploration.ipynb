```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("ahsanaseer/top-rated-tmdb-movies-10k")

print("Path to dataset files:", path)
```

```python
import pandas as pd

movies = pd.read_csv(f"{path}/top10K-TMDB-movies.csv")
```

```python
movies
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#### DATA CLEANUP - remove the movies with no genre or overview
movies[movies["overview"].isna()]
```

```python
movies = movies.dropna(subset=["overview", "genre"])
```

```python
movies[movies["overview"].isna()]
```

```python
movies[movies["genre"].isna()]
```

```python
movies
```

```python
# check duplicate movie titles
movies[movies["title"].duplicated(keep=False)].sort_values("title")
```

```python
# check duplicate movie ids
movies[movies["id"].duplicated(keep=False)].sort_values("title")
```

```python
# check if overviews are meaningful
movies["overview_word_count"] = movies["overview"].str.split().str.len()
movies[movies["overview_word_count"] < 10][
    ["title", "overview", "overview_word_count"]
].sort_values("overview_word_count")
```

```python
movies = movies[movies["overview"].str.lower() != "a comedy show"]
```

```python
movies[movies["overview"].str.lower() == "a comedy show"]
```
